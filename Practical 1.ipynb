1. Sum of two numbers

```python
num1 = 5
num2 = 10
sum = num1 + num2
print("The sum of", num1, "and", num2, "is", sum)
```

2. Factorial of a Number

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

num = 5
print("The factorial of", num, "is", factorial(num))
```

3. Fibonacci series

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

terms = 10
print("Fibonacci series up to", terms, "terms:")
for i in range(terms):
    print(fibonacci(i))
```

4. Check if is prime

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = 17
if is_prime(num):
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

5. Reverse a string

```python
string = "Hello, World!"
reversed_string = string[::-1]
print("Reversed string:", reversed_string)
```

6. Check if is Palindrome

```python
def is_palindrome(string):
    if string == string[::-1]:
        return True
    return False

word = "radar"
if is_palindrome(word):
    print(word, "is a palindrome")
else:
    print(word, "is not a palindrome")
```

7. Sum of digits

```python
def sum_of_digits(n):
    sum = 0
    while n > 0:
        sum += n % 10
        n //= 10
    return sum

num = 12345
print("Sum of digits in", num, "is", sum_of_digits(num))
```

8. Check if is an armstrong number

```python
def is_armstrong(n):
    num = n
    sum = 0
    while num > 0:
        digit = num % 10
        sum += digit ** 3
        num //= 10
    if n == sum:
        return True
    return False

num = 153
if is_armstrong(num):
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

9. Count no. of vowels

```python
def count_vowels(string):
    vowels = "aeiouAEIOU"
    count = 0
    for char in string:
        if char in vowels:
            count += 1
    return count

word = "Hello, World!"
print("Number of vowels in", word, "is", count_vowels(word))
```

10. Check if Leap Year

```python
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False

year = 2024
if is_leap_year(year):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```
